```python
from karpathy_series.micrograd.assignment import Assignment
from karpathy_series.micrograd.calculation import GraphValuation
from karpathy_series.micrograd.drawing import Drawing
from karpathy_series.micrograd.value import ValueGraph
```

```python
G = ValueGraph()
a, b, c, f = G["a", "b", "c", "f"]
e = a * b | "e"
d = e + c | "d"
L = d * f | "L"
drawing = Drawing().draw_value_graph(G)
drawing
```

```python
gv_1 = GraphValuation.initialize(Assignment.create(G, {a: 1, b: 2, c: 3, f: 5}))
_ = gv_1.forward()
_ = gv_1.backward()
drawing = Drawing().draw_graph_valuation(gv_1)
drawing
```
