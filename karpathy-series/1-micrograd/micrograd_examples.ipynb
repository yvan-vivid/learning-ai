```python
from micrograd.value import ValueGraph
from micrograd.assignment import Assignment
from micrograd.calculation import GraphValuation
from micrograd.drawing import Drawing
```

```python
G = ValueGraph()
a, b, c, f = G['a', 'b', 'c', 'f']
e = a*b | 'e'
d = e + c | 'd'
L = d * f | 'L'
Drawing().draw_value_graph(G)
```

```python
gv_1 = GraphValuation.initialize(Assignment.create(G, {
    a: 1,
    b: 2,
    c: 3,
    f: 5
}))

gv_1.forward()
gv_1.backward()
Drawing().draw_graph_valuation(gv_1)
```
